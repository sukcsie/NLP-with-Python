```
import re
reader = csv.reader(open('all.csv', 'r',newline='\n'))
d = {}
for k,v in reader:
  if re.search('^[a-zA-Z]+$', k):
    d[k] = int(v)
    #print(k, "\t", d[k])

```

```
#Generating wordcloud. Relative scaling value is to adjust the importance of a frequency word.
#See documentation: https://github.com/amueller/word_cloud/blob/master/wordcloud/wordcloud.py
wordcloud = WordCloud(width=1000,height=500, max_words=50,relative_scaling=0.5,
                      normalize_plurals=False, background_color=None, mode="RGBA").generate_from_frequencies(d)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
